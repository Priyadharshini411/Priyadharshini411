```python
num1 = int(input("Enter first number: "))
num2 = int(input("Enter second number: "))

# Addition
print(f"Addition: {num1} + {num2} = {num1 + num2}")

# Subtraction
print(f"Subtraction: {num1} - {num2} = {num1 - num2}")

# Multiplication
print(f"Multiplication: {num1} * {num2} = {num1 * num2}")

# Division and Modulus only if second number is not zero
if num2 != 0:
    print(f"Division: {num1} / {num2} = {num1 / num2}")
    print(f"Modulus: {num1} % {num2} = {num1 % num2}")
else:
    print("Division and Modulus not possible (division by zero)")

# Exponentiation (Power)
print(f"Exponentiation: {num1} ** {num2} = {num1 ** num2}")
```

```python
import numpy as np   # Import NumPy library

# Create a list with 10 numbers
numbers = [10, 25, 40, 5, 60, 75, 90, 15, 35, 50]

print("Original List:", numbers)

# Print length of the list
print("Length of the list:", len(numbers))

# Find maximum and minimum value
print("Maximum value:", max(numbers))
print("Minimum value:", min(numbers))

# Add a new element
numbers.append(100)
print("After adding 100:", numbers)

# Remove one element
numbers.remove(5)
print("After removing 5:", numbers)

# Sort in ascending order
numbers.sort()
print("Ascending Order:", numbers)

# Sort in descending order
numbers.sort(reverse=True)
print("Descending Order:", numbers)

# Convert list into NumPy array
array = np.array(numbers)
print("NumPy Array:", array)

# Calculate Mean, Median and Standard Deviation
print("Mean:", np.mean(array))
print("Median:", np.median(array))
print("Standard Deviation:", np.std(array))
```

```python

student = {
    "name": "Priya",
    "age": 20,
    "course": "AI & DS",
    "marks": 88
}

# Print each key-value using loop
print("Student Details:")
for key, value in student.items():
    print(key, ":", value)

# Add new key called grade
student["grade"] = "A"
print("\nUpdated Dictionary:", student)

# Create a set with duplicate values
courses = {"Python", "Data Science", "AI", "Python"}
print("\nUnique Courses Set:", courses)

# Create two sample sets
set1 = {"Python", "Java", "C"}
set2 = {"Python", "AI", "Data Science"}

# Perform set operations
print("\nUnion:", set1.union(set2))
print("Intersection:", set1.intersection(set2))
print("Difference:", set1.difference(set2))
```
